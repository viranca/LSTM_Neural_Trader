```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from functools import reduce
%matplotlib inline
```

```python
#Import OANDA Data
df_eurusd = pd.read_csv('Data_OANDA/EURUSD12_11_2018.csv')
df_eurgbp = pd.read_csv('Data_OANDA/EURGBP12_11_2018.csv')
df_gbpusd = pd.read_csv('Data_OANDA/GBPUSD12_11_2018.csv')
```

```python
#Remove columns Time from OANDA data
dfs_OANDA = [df_eurusd, df_eurgbp, df_gbpusd]
dfs_OANDA_new =[]
for a in dfs_OANDA:
    a.drop(a.columns[[1]],axis=1, inplace=True)
    dfs_OANDA_new.append(a)
```

```python
#Merge OANDA Data    
dfm_OANDA = reduce(lambda left,right: pd.merge(left,right,on='Date'), dfs_OANDA_new)
```

```python
#Change date format from OANDA
new_date = []
for day in dfm_OANDA['Date']:
    newday = ""
    newdaay = []
    for letter in day:
        letter = letter.replace(".","-")
        newday = newday + letter
    newdaay.append(newday)
    new_date.append(newday)
dfm_OANDA['Date'] = new_date
```

```python
#Import Yahoo Data
df_aex = pd.read_csv('Data_Yahoo/AEX.csv')
```

```python
#Merge Yahoo Data into OANDA merged Data
dfs = [dfm_OANDA, df_aex]
dfm = reduce(lambda left,right: pd.merge(left,right,on='Date'), dfs)
```
